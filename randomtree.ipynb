```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# Loading the data


```python
df = sns.load_dataset('penguins')
df.head()
```

```python
df.shape
```

```python
df.info()
```

# Dropping the null values


```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum() #null values have been dropped
```

```python
df.sex.unique()
```

# **Feature Engineering**

```python
#using one code hot encoding transforming the categorical data into numeric
pd.get_dummies(df['sex']).head()
```

```python
sex = pd.get_dummies(df['sex'], drop_first=True)
sex.head()
```

```python
# applying one hot encoding to island
df.island.unique()
```

```python
pd.get_dummies(df['island']).head()
```

```python
island = pd.get_dummies(df['island'], drop_first=True)
island.head(5)
```

# Concatenating the above two data frames to the original data

```python
new_data = pd.concat([df, island, sex], axis = 1)
```

```python
new_data.head()
```

```python
new_data.drop(['sex', 'island'], axis=1, inplace=True) # dropping the repeated column
```

```python
new_data.head()
```

#Creating seperte target variables

```python
Y = new_data.species
Y.head()
```

```python
Y.unique()
```

```python
Y =Y.map({'Adelie':0, 'Chinstrap':1, 'Gentoo':2})
Y.head()
```

```python
new_data.drop('species', inplace =True, axis=1) # dropping the target value species
```

```python
new_data.head()
```

```python
X= new_data
```

# Splitting the dataset into training and testing data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print('X_train:', X_train.shape)
print('X_test:', X_test.shape)
print('Y_train:', y_train.shape)
print('Y_test:,', y_test.shape)
```

# Training Random Forest Classification on training set


```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=5, criterion='entropy', random_state=0)
classifier.fit(X_train, y_train)
```

# Predicting the results

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, accuracy_score
```

```python
cm=confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
